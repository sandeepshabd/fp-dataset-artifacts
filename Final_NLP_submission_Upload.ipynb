```python
from datasets import load_dataset
import numpy as np
import re

```

```python
!pwd
```

```python

squad = load_dataset('squad', split='validation')

squad_question_lengths = np.zeros((squad.shape[0],))
squad_overlap = np.zeros((squad.shape[0],))

for i in range(len(squad)):
    squad_question_lengths[i] = len(list(set(re.split('\W+', squad[i]['question']))))
    squad_overlap[i] = len(list(set(re.split('\W+', squad[i]['question'])) & set(re.split('\W+', squad[i]['context']))))


np.savetxt('/content/fp-dataset-artifacts/squad_eval_question_lengths.txt', squad_question_lengths)
np.savetxt('/content/fp-dataset-artifacts/squad_eval_set_overlap.txt', squad_overlap)
```

```python

adversial = load_dataset('adversarial_qa', 'adversarialQA', split='validation')

adversial_question_lengths = np.zeros((adversial.shape[0],))
adversial_overlap = np.zeros((adversial.shape[0],))

for i in range(len(adversial)):
    adversial_question_lengths[i] = len(list(set(re.split('\W+', adversial[i]['question']))))
    adversial_overlap[i] = len(list(set(re.split('\W+', adversial[i]['question'])) & set(re.split('\W+', adversial[i]['context']))))


np.savetxt('/content/fp-dataset-artifacts/adversial_eval_question_lengths.txt', adversial_question_lengths)
np.savetxt('/content/fp-dataset-artifacts/adversial_eval_set_overlap.txt', adversial_overlap)
```

```python
questions_lengths_squad = np.loadtxt('/content/fp-dataset-artifacts/squad_eval_question_lengths.txt')
questions_lengths_squad_mean = np.mean(squad_question_lengths)

print(questions_lengths_squad_mean)
```

```python
questions_lengths_squad = np.loadtxt('/content/fp-dataset-artifacts/squad_eval_question_lengths.txt')
squad_question_overlap = np.loadtxt('/content/fp-dataset-artifacts/squad_eval_set_overlap.txt')

# mean percentage of overlap
squad_overlap = squad_question_overlap / questions_lengths_squad
mean_overlap = np.mean(squad_overlap) * 100
print(mean_overlap)
```

```python
questions_lengths_adversial = np.loadtxt('/content/fp-dataset-artifacts/adversial_eval_question_lengths.txt')
questions_lengths_adversial_mean = np.mean(adversial_question_lengths)

print(questions_lengths_adversial_mean)
```
